# Python Visualizations: `matplotlib` and `seaborn`

`matplotlib` is the classic python visualization. It's well-supported and well-documented, and was initially released way back in 2003. It supports a wide range of visualization types, all the way from the traditional scatter plot and line graph, to more complex contour plots. Basic, static charts for scientific papers can "easily" be gerenated using `matplotlib`.

Recently, there has been an abundance of new visualization tools. Some are particularly useful in generated more advanced visualizations, prettier visualizations, or dynamic visualizations.

`seaborn`, which we'll also use in this course is build on-top of `matplotlib`. It tends to generate prettier graphics, can also generate other graphic types such as heat maps, and is compatible with the Python dataframe library `pandas` that we'll also utilize in this course.

## Other python visualization tools

`ggplot` has syntax based on the R programming language's `ggplot2` library, and is widely used.

`D3` is a popular visualization tool in JavaScript for producing interactive web graphics. 

`Bokeh` is an interactive visualization library written in Python.

# "Simple" Plots / Single Line Graph

Visualizations in `matplotlib` are built-up, step by step. 

Successive commands are often used to customize charts, and can change the axes labels, line styles, etc.

````python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
````

## Histogram

Note that `import` statements _are_ allowed within notebook cells.

In this example, I've removed the "shifting each bar left" code from the Grus example.

````python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10

histogram = Counter(decile(grade) for grade in grades)

plt.bar(histogram.keys(),                  # grab each bar's name
        histogram.values(),                # give each bar its correct height
        width=8)                           # give each bar a width of 8

plt.axis([-5, 105, 0, 5])                  # x-axis from -5 to 105,
                                           # y-axis from 0 to 5
    
plt.xticks([10 * i for i in range(11)])    # x-axis labels at 0, 10, ..., 100

plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
````

### Changing y-axis scale

In this example, I removed the uncessary `0.8` width and call to `plt.ticklabel_format`.

This function definition promotes code reuse.

````python
def make_chart_misleading_y_axis(mislead=True):

    mentions = [500, 505]
    years = [2013, 2014]

    plt.bar(years, mentions)
    plt.xticks(years)
    plt.ylabel("# of times I heard someone say 'data science'")

    if mislead:
        # misleading y-axis only shows the part above 500
        plt.axis([2012.5,2014.5,499,506])
        plt.title("Look at the 'Huge' Increase!")
    else:
        plt.axis([2012.5,2014.5,0,550])
        plt.title("Not So Huge Anymore.")
    plt.show()
````

````python
make_chart_misleading_y_axis(mislead=True)
````

````python
make_chart_misleading_y_axis(mislead=False)
````

## Generating two charts side by side

`matplotlib` can generate plots within plots, but the syntax has never been intuitive to me.

I definitely need to consult the API when I start using `plt.subplots`.

````python
mentions = [500, 505]
years = [2013, 2014]

fig, axs = plt.subplots(nrows=1, ncols=2)

plt.tight_layout(pad=2, w_pad=10, h_pad=0)

axs[0].bar(years, mentions)
axs[0].set_xticks(years)
axs[0].set_ylabel("# of times I heard someone say 'data science'")
axs[0].axis([2012.5,2014.5,499,506])
axs[0].set_title("Look at the 'Huge' Increase!")

axs[1].bar(years, mentions)
axs[1].set_xticks(years)
axs[1].set_ylabel("# of times I heard someone say 'data science'")
axs[1].axis([2012.5,2014.5,0,550])
axs[1].set_title("Not So Huge Anymore.")

fig.suptitle('Huge vs not so huge', y=1.10, fontsize=20)
````

# Line Graphs

A good choice for showing _trends_.

````python
variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]

xs = range(len(variance))

# we can make multiple calls to plt.plot
# to show multiple series on the same chart
plt.plot(xs, variance,     'g-',  label='variance')    # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')      # red dot-dashed line
plt.plot(xs, total_error,  'b:',  label='total error') # blue dotted line

# because we've assigned labels to each series
# we can get a legend for free
# loc=9 means "top center"
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
````

# Scatterplots

A good choice for visualizing the realtionship between two paired sets of data.

````python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count), # put the label with its point
                 xytext=(5, -5), # but slightly offset
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
````

## Changing the Scale

````python
def make_chart_scatterplot_axes(equal_axes=False):
    test_1_grades = [ 99, 90, 85, 97, 80]
    test_2_grades = [100, 85, 60, 90, 70]

    plt.scatter(test_1_grades, test_2_grades)
    plt.xlabel("test 1 grade")
    plt.ylabel("test 2 grade")

    if equal_axes:
        plt.title("Axes Are Comparable")
        plt.axis("equal")
    else:
        plt.title("Axes Aren't Comparable")

    plt.show()
````

````python
make_chart_scatterplot_axes(equal_axes=False)
````

````python
make_chart_scatterplot_axes(equal_axes=True)
````

## Reading in the `iris.data` file

(Note that there are some File IO differences between Python2 and Python3. Below is for Python3.)

````python
import csv

with open('iris.data') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        print(', '.join(row))
````

Saving data into our own python data structure. We'll make a separate list for each feature.

````python
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
species = []

import csv

with open('iris.data') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        if row:    # checking if row is not the empty list, which is returned for the last row
            sepal_length.append(float(row[0]))    # string to float conversion function necessary!
            sepal_width.append(float(row[1]))
            petal_length.append(float(row[2]))
            petal_width.append(float(row[3]))
            species.append(row[4])
            
print("%d rows loaded" % len(sepal_length))
````

### Percentiles

Grus code!

Function that generalizes the computation of a percentile. Not in the statistics module.

````python
def quantile(x,p):
    """returns the pth-percentile value in x"""
    p_index = int(p * len(x))
    return sorted(x)[p_index]
````

````python
print(quantile(sepal_length, 0.50))
print(statistics.median(sepal_length))
````

### Dispersion, measure of spread

````python
def data_range(x):
    return max(x) - min(x)
````

````python
def interquartile_range(x):
    return quantile(x, 0.75) - quantile(x, 0.25)
````

````python
print(data_range(petal_length))
print(interquartile_range(petal_length))
````
